```python
#visualization
df.groupby('State')['TotalAmount'].sum().plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Revenue')
plt.title('Revenue by State')
plt.show()
```

```python
#category wise Sales
df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False).plot(kind='pie')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.title('Revenue by Category')
plt.show()
```

```python
#ANSWER 7: month recorded the highest sales and profit? Is there any noticeable sales trend
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

monthly_sales = df.groupby(df['OrderDate'].dt.month_name())[['TotalAmount','Revenue']].sum()

print(monthly_sales.sort_values(by='TotalAmount', ascending=False))

#August      7890598.37  7.358047e+06 =month recorded the highest sales and profit
#Higher sales during certain months may be influenced by seasonal demand, promotional campaigns, or festive shopping periods
```

```python
# ANSWER 8: categories or products with high sales but low profit. What could be the possible reason: 
product_analysis = df.groupby('ProductName')[['TotalAmount','Revenue']].sum()

product_analysis['Difference'] = product_analysis['TotalAmount'] - product_analysis['Revenue']

product_analysis.sort_values(by='Difference', ascending=False).head(10)

#The analysis shows that some products generate high sales but experience lower revenue due to larger discounts offered to customers.
```
